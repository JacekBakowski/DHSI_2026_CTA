### LOADING THE DATASETS

```python
PATH_100_NOVELS = '../corpora/100_English_Novels/raw/corpus'
PATH_100_NOVELS_LEMMA = '../corpora/100_English_Novels/lemmatized/corpus'
PATH_REF = '../corpora/Wordlists/bnc_coca_word.tsv'
PATH_REF_LEMMA = '../corpora/Wordlists/bnc_coca_lemma.tsv'
PATH_SMALL_COLLECTION = '../corpora/A_Small_Collection_of_British_Fiction/raw/corpus'
PATH_SMALL_COLLECTION_LEMMA = '../corpora/A_Small_Collection_of_British_Fiction/lemmatized/corpus'


freq_100 = get_corpus_frequency_list (PATH_100_NOVELS)
freq_100_lemma = get_corpus_frequency_list (PATH_100_NOVELS_LEMMA) 
freq_small = get_corpus_frequency_list (PATH_SMALL_COLLECTION)
freq_small_lemma = get_corpus_frequency_list (PATH_SMALL_COLLECTION_LEMMA)
freq_ref = get_corpus_frequency_list (PATH_REF, wordlist = True)
freq_ref_lemma = get_corpus_frequency_list (PATH_REF_LEMMA, wordlist = True)
```

### Log-Likelihood Keyness
#### 100 Novels (raw)

```python
ll_100 = calculate_keyness (freq_100, freq_ref)
print ('100 NOVELS (RAW)\n===================\n')
for ind, (word, ll) in enumerate (ll_100[:50], start = 1):
    print (f'{ind:<2}: {word:<20}{ll}')
```

#### 100 Novels (lemmatized)

```python
ll_100_lemma = calculate_keyness (freq_100_lemma, freq_ref_lemma)
print ('100 NOVELS (LEMMATIZED)\n===================\n')
for ind, (word, ll) in enumerate (ll_100_lemma[:50], start = 1):
    print (f'{ind:<2}: {word:<20}{ll}')
```

#### A Small Collection (lemmatized)

```python
ll_small_lemma = calculate_keyness (freq_small_lemma, freq_ref_lemma)
print ('SMALL COLLECTION (LEMMATIZED)\n===================\n')
for ind, (word, ll) in enumerate (ll_small_lemma[:50], start = 1):
    print (f'{ind:<2}: {word:<20}{ll}')
```

#### 100 Novels vs Small Collection

```python
ll_100_vs_small = calculate_keyness (freq_100_lemma, freq_small_lemma)
print ('SMALL COLLECTION (LEMMATIZED)\n===================\n')
for ind, (word, ll) in enumerate (ll_100_vs_small[:50], start = 1):
    print (f'{ind:<2}: {word:<20}{ll}')
```
