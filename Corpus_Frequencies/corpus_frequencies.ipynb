```python
PATH_100_NOVELS = '../corpora/100_English_Novels/raw/corpus'
PATH_100_NOVELS_LEMMA = '../corpora/100_English_Novels/lemmatized/corpus'
PATH_REF = '../corpora/Wordlists/bnc_coca_word.tsv'
PATH_REF_LEMMA = '../corpora/Wordlists/bnc_coca_lemma.tsv'
```

```python
freq_100 = get_corpus_frequency_list (PATH_100_NOVELS)
freq_100_lemma = get_corpus_frequency_list (PATH_100_NOVELS_LEMMA)
freq_ref = get_corpus_frequency_list (PATH_REF, wordlist = True)
freq_ref_lemma = get_corpus_frequency_list (PATH_REF_LEMMA, wordlist = True)

TOTAL = {
    '100': get_total_counts (freq_100),
    'ref': get_total_counts (freq_ref),
}
```

```python
print (f'{"Vampire (100 Novels):":<40}{freq_100["vampire"]}')
print (f'{"Vampire (100 Novels - lemmatized):":<40}{freq_100_lemma["vampire"]}')
print (f'{"Vampire (BNC/COCA):":<40}{freq_ref["vampire"]}')
print (f'{"Vampire (BNC/COCA) - lemmatized:":<40}{freq_ref_lemma["vampire"]}')
```

```python
words = {
    'castle': 751,
    'ruin': 560,
    'ghost': 650,
    'vampire': 12,
}
```

```python
freq_100['vampire']
```

```python
freq_100_lemma['vampire']
```

```python


words_relative = {w: relative_frequency (words[w], TOTAL['100']) for w in words}
ref_relative = {w: relative_frequency (freq_ref.get (w, 0), TOTAL['ref']) for w in words}

print (f'{"word":<20}{"frequency (per M)":<20}{"reference (per M)":<20}{"ratio"}')
for w in words_relative:
    f = words_relative[w]
    r = ref_relative[w]
    ratio = f / r
    print (f'{w:<20}{words_relative[w]:<20.1f}{ref_relative[w]:<20.1f}{ratio:.2f}')
```

```python
plot_word_frequency (words_relative)

```

```python
plot_word_frequency (words_relative, ref_relative)
```
